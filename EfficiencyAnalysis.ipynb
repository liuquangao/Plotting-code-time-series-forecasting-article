# Efficiency Analysis


## 1. As the Sequence Length Increases

### 1.1 GpuMemory(MB)

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
series_length = [96, 192, 384, 768, 1536, 3072]
gpu_memory = {
    "MSTVI": [1950, 1970, 1972, 1990, 2082, 2172],
    "iTransformer": [2540,2540,2570,2616,2632,2706],
    "TimesNet": [4178,4258,4784,6416,9068,11488],
    "FEDformer": [2636,2746,3082,4046,6476,11294],
    "Autoformer": [2676,3336,4796,8036,15184,30000],
    "Informer": [1622,1976,2254,4144,7550,15740],
}

# Colors corresponding to each model
colors = {
    "MSTVI": "blue",
    "iTransformer": "cyan",
    "TimesNet": "orange",
    "RLinear": "red",
    "FEDformer": "purple",
    "Autoformer": "brown",
    "Informer": "green",
}

# Create an evenly spaced x-axis
x_positions = np.arange(len(series_length))

# Plotting
plt.figure(figsize=(10, 6))
for model, memory in gpu_memory.items():
    plt.plot(x_positions, memory, marker='o', color=colors[model], label=model, linewidth=3)

plt.xlabel('Series Length', fontsize=15)
plt.ylabel('GPU Memory (MB)', fontsize=15)
plt.title('GPU Memory by Series Length', fontsize=20)
plt.legend(fontsize=15)

# Set x-ticks to be the original series_length values at even intervals
plt.xticks(x_positions, [str(x) for x in series_length])

# Remove grid lines
plt.grid(False)

plt.savefig("gpu_memory_by_series_length.png", bbox_inches='tight',dpi=600)

plt.tight_layout()
plt.show()
```
